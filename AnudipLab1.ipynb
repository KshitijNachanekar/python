### Q1 Python program to check leap year

```python
year = int(input("Enter the Year: "))
if ( year%4==0 and year%100!=0 ) or (year%400==0):
  print(year," is Leap Year")
else:
  print(year, " is not Leap Year")
```

4. A toy vendor supplies three types of toys: Battery Based Toys, Key-based
Toys, and Electrical Charging Based Toys. The vendor gives a discount of
10% on orders for battery-based toys if the order is for more than Rs. 1000.
On orders of more than Rs. 100 for key-based toys, a discount of 5% is
given, and a discount of 10% is given on orders for electrical charging based
toys of value more than Rs. 500. Assume that the numeric codes 1,2 and 3
are used for battery based toys, key-based toys, and electrical charging based
toys respectively. Write a program that reads the product code and the order
amount and prints out the net amount that the customer is required to pay
after the discount.

```python
print("1) Battery Based Toys\n2) Key-based Toys\n3) Electrical Charging Based Toys")
code = int(input("Enter the Toy Code: "))
amount = int(input("Enter the Order Amount"))
if code==1:
  if amount>=1000:
    disc = 0.1*amount
    print("Total Discounted Bill",amount-disc)
  else:
    print("Total Bill",amount)

elif code==2:
  if amount>=100:
    disc = 0.05*amount
    print("Total Discounted Bill",amount-disc)
  else:
    print("Total Bill",amount)

elif code==3:
  if amount>=500:
    disc = 0.1*amount
    print("Total Discounted Bill",amount-disc)
  else:
    print("Total Bill",amount)

else:
  print("Invalid Code")
```

## Write a python program to reverse a number using a while loop.

```python
num = int(input("Enter the Number: "))
reversed = 0

while num != 0:
    digit = num % 10
    reversed = reversed * 10 + digit
    num //= 10

print("Reversed Number: " + str(reversed))
```

## Write a python program to check whether a number is palindrome or not?

```python
num = input("Enter the Number: ")
reversed = str(num)[::-1]
if reversed == num:
  print("The number is Palindrome")
else:
  print("The number is not Palindrome")
```

## Write a python program finding the factorial of a given number using a while loop.

```python
num = int(input("Enter the number: "))
fact = 1
while num > 1:
    fact = fact * num
    num = num - 1
print("Factorial of", num, "is", fact)
```

## Accept numbers using input() function until the user enters 0. If user input 0 then break the while loop and display the sum of all the numbers.

```python
num = int(1)
sum1 = int(0)
while num != 0:
  sum1 = sum1 + num
  num = int(input("Enter the Number: "))
print("The sum is: ", sum1-1)
```

## Python program to check if a given number is an Armstrong number

```python
num = int(input("Enter The Number: "))
num2 = str(num)
n = len(num2)
sum1 = 0

for digit in num2:
    sum1 += int(digit) ** n

if sum1 == num:
    print(num, "is an Armstrong number")
else:
    print(num, "is not an Armstrong number")
```

## Python program to get the Fibonacci series between 0 to 50

```python
n=11
a, b = 0, 1
if n <= 0:
  print([])
elif n == 1:
  print(0)
else:
  sequence = [0, 1]
  for _ in range(2, n):
    next_num = a + b
    sequence.append(next_num)
    a, b = b, next_num
  print(sequence)
```

## Python program to check the validity of password input by users

```python
pass1 = str(input("Enter the Password: "))
l = u = p = d = 0

if len(pass1) >= 8:
    for ch in pass1:
        if ch.islower():
            l += 1
        if ch.isupper():
            u += 1
        if ch.isdigit():
            d += 1
        if ch in "_@$":
            p += 1

if l >= 1 and u >= 1 and p >= 1 and d >= 1 and l+u+d+p == len(pass1):
    print("Valid Password")
else:
    print("Invalid Password")
```
