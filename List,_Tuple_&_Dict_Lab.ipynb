```python
a = [50, 60, 10, 20, 30]
largest , smallest = 0, 100
for i in a:
  if i>smallest:
    if i>largest:
      largest = i
  else:
    smallest = i
print("The largest number is:", largest, "and the Smallest is:", smallest)
```

## 3. Write a Python program to find duplicate values from a list and display those.

```python
a = [50, 60, 10, 20, 30, 20, 30, 50]
freq = dict()
for i in a:
  if i not in freq:
    if a.count(i)>=2:
      freq.setdefault(i,a.count(i))
print(freq.keys())
```

## 5. Write a Python program to traverse a given list in reverse order, and print the elements with the original index.
Original list:
['red', 'green', 'white', 'black']

Traverse the said list in reverse order:

black
white
green
red

```python
orig = ['red', 'green', 'white', 'black']
reversed = orig[::-1]
j = len(orig)
for i in reversed:
  j-=1
  print(i,j)
```

## 4.Write a Python program to get the key, value and item in a dictionary.

```python
input_dict = {1: 10, 2: 20, 3:None, 4: 40, 5: None, 6: 60}
for i in list(input_dict.keys()):
    if input_dict[i] is None:
        input_dict.pop(i)
print("Dictionary with Empty Items Dropped:")
print(input_dict)
```

## 3. Write a Python program to calculate the sum of the numbers in a given tuple.

**Input:** tuples_list = [(1, 2), (3, 4), (5, 6)]

```python
tuples_list = [(1, 2), (3, 4), (5, 6)]
sum = 0
for i in tuples_list:
  for j in i:
    sum +=j
print("The sum of the values in the tuple is:",sum)
```

## 1. Write a Python program to Get Only unique items from two sets.

**Input:**
set1 = {10, 20, 30, 40, 50}
set2 = {30, 40, 50, 60, 70}

**Output:**
{70, 40, 10, 50, 20, 60, 30}

```python
set1 = {10, 20, 30, 40, 50}
set2 = {30, 40, 50, 60, 70}
set3 = set1.union(set2)
print(set3)
```

## 2. Write a Python program to Return a set of elements present in Set A or B, but not both.

**Input:**
set1 = {10, 20, 30, 40, 50}
set2 = {30, 40, 50, 60, 70}

**Output:**
{20, 70, 10, 60}

```python
set1 = {10, 20, 30, 40, 50}
set2 = {30, 40, 50, 60, 70}
set3 = set1.symmetric_difference(set2)
print(set3)
```

## 3. Write a Python program to Check if two sets have any elements in common. If yes, display the common elements.

**Input:**
set1 = {10, 20, 30, 40, 50}
set2 = {60, 70, 80, 90, 10}

**Output:**
{10}

```python
set1 = {10, 20, 30, 40, 50}
set2 = {60, 70, 80, 90, 10}
set3 = set1.intersection(set2)
print(set3)
```
