## **1. Write a Python program to count the occurrences of each word in a given sentence**
### string = “To change the overall look of your document. To change the look available in the gallery”

```python
a = "To change the overall look of your document. To change the look available in the gallery"
a = a.lower().strip().split()

num = {}
for word in a:
  if word in num:
    num[word] += 1
  else:
    num[word] = 1
print(num)
```

## **3. Write a Python program to reverse words in a string**
## String = “Deeptech Python Training”

```python
a = "Deeptech Python Training"
for line in a.split('\n'):
	print(' '.join(line.split()[::-1]))
```

## **4. Write a Python program to count and display the vowels of a given text**
## String=”Welcome to python Training”

```python
a = "Welcome to python Training"
a = a.lower().strip()
vowels = "aeiou"
count = 0
for i in a:
  if i in vowels:
    count+=1
print("Vowel count is:", count)
```

##**1. Write a Python program to Count all letters, digits, and special symbols from the given string**
## Input = “P@#yn26at^&i5ve”
## Output: Chars = 8 Digits = 2 Symbol = 3

```python
a = "P@#yn26at^&i5ve"
alphabets=0
digits=0
specialChars=0
for i in a:
  if i.isalpha():
    alphabets+=1
  elif i.isdigit():
    digits+=1
  else:
    specialChars+=1
print("alphabets =",alphabets,"digits =",digits,"specialChars =",specialChars)
```

## **2. Write a Python program to remove duplicate characters of a given string.**
## Input = “String and String Function”
## Output: String and Function

```python
a = "String and String Function"
a = a.split()
result=""
for i in a:
  if i not in result:
    result = result+" "+i
print(result)
```

## **3. Write a Python program to count Uppercase, Lowercase, special character and numeric values in a given string**

Input = “Hell0 W0rld ! 123 * # welcome to pYtHoN”

Output:
UpperCase : 5
LowerCase : 18
NumberCase : 5
SpecialCase : 11

```python
a = "Hell0 W0rld ! 123 * # welcome to pYtHoN"
upper = 0
lower = 0
num = 0
special = 0
for i in a:
  if i.isalpha():
    if i.islower():
      lower+=1
    else:
      upper+=1
  elif i.isnumeric():
    num +=1
  else:
    special +=1
print("UpperCase :",upper,"\nLowerCase :",lower, "\nNumberCase :", num,"\nSpecialCase :", special)
```
