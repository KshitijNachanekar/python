## Slicing Syntax
array[start:end]

```python
arr = np.array([10,20,30,40,50])
print(arr[1:4])
print(arr[:3])
print(arr[2:])
print(arr[-3:])
```

# Numpy Functions

```python
# arrange()
arr = np.arange(1,6)
print(arr)
```

```python
#Zeros()
print(np.zeros(5))
```

```python
#ones()
print(np.ones(5))
```

```python
# sum(), mean(),max(),min()
marks = np.array([70,80,90,60])

print("sum:",np.sum(marks))
print("Average:",np.mean(marks))
print("Max:",np.max(marks))
print("min:",np.min(marks))
```

# 2D Numpy Array

```python
import numpy as np
arr=np.array([
    [10,20,30,14],
    [40,50,60,24],
    [70,80,90,34],
    [11,22,33,44]
])
print(arr)
print(arr.shape)
print(arr[0,2]) #indexing
print(arr[1:3,1:3]) #slicing
print(arr+2)
print(arr*2)
```

```python
#axis=0 / column
#axis=1 / row
```

```python
marks= np.array([
    [10,20,30],
    [40,50,60],
    [70,80,90]
])
print("average marks",np.mean(marks,axis=1))
print("average marks",np.mean(marks,axis=0))
```
